```python
from faker import Faker
fake= Faker()
names = [fake.name().split() for i in range(100)]
# to normalize the date make sure that each list  in names must have only two elements
names = [name for name in names if len(name) == 2]
```

```python
names
```

```python
import sqlite3
connection = sqlite3.connect("sample.db")
```

```python
insert_query= 'INSERT into people(name , surmane) VALUES (?,?)'
cursor = connection.cursor()
for name in names:
    cursor.execute(insert_query, name)
connection.commit()
```
